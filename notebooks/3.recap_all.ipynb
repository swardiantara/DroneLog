```python
import os
import pandas as pd
recap_baseline = pd.read_excel('experiments/recap/baseline/recap_overall_baseline.xlsx')
recap_multitask = pd.read_excel('experiments/recap/multitask/recap_overall_multitask.xlsx')
recap_all = pd.concat([recap_baseline, recap_multitask], ignore_index=True)
```

```python
output_dir = os.path.join('experiments', 'recap')
recap_all.to_excel(os.path.join(output_dir, 'final-recap-all.xlsx'), index=False)
```
